```python
def demo_ss_format(g):
    control.reset_defaults()
    print(repr(g))
    print('default: partitioned, fmt="".3g"')
    display(g)
    control.set_defaults('statesp', latex_num_format='.5g')
    print('partitioned, fmt="".5g"')
    display(g)
    control.reset_defaults()
    control.set_defaults('statesp', latex_repr_type='separate')
    print('separate, fmt="".3g"')
    display(g)
    control.set_defaults('statesp', latex_num_format='.5g')
    print('separate, fmt="".5g"')
    display(g)
```

```python
demo_ss_format(g1)
```

```python
demo_ss_format(g6)
```

```python
demo_ss_format(g7)
```
